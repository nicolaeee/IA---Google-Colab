```python
#Afișarea data-set
import pandas as pd


#Citirea fișierului
file_path = "/content/freelancer_earnings_bd.csv"
df = pd.read_csv(file_path, encoding="ISO-8859-1")

#Afișarea primelor rânduri
df.head()
```

```python
#Creare matrice din datele alese
import numpy as np

# Selectăm coloanele numerice pentru analiză
selected_data = df[["Earnings_USD", "Hourly_Rate", "Job_Success_Rate", "Client_Rating"]]

# Convertim într-o matrice numpy
data_matrix = selected_data.to_numpy()

print("Matricea numerică (primele 5 rânduri):")
print(data_matrix[:5])
```

```python
# Media
average_values = selected_data.mean()
print("\n📊 Media valorilor:")
print(average_values)
```

```python
# Mediana
median_values = selected_data.median()
print("\n📊 Mediana valorilor:")
print(median_values)
```

```python
# Valoarea cea mai frecventa
mode_values = selected_data.mode().iloc[0]
print("\n📊 Modul (valoarea cea mai frecventă):")
print(mode_values)
```

```python
#Range min max
range_values = selected_data.max() - selected_data.min()
print("\n📊 Range (domeniul valorilor):")
print(range_values)
```

```python
#Deviația standard
std_dev = selected_data.std()
print("\n📊 Deviația standard:")
print(std_dev)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogramă
selected_data.hist(figsize=(10, 8), bins=15, edgecolor='black')
plt.suptitle("📊 Histogramă pentru fiecare coloană numerică")
plt.tight_layout()
plt.show()

# Box plot
plt.figure(figsize=(10, 6))
sns.boxplot(data=selected_data)
plt.title("📦 Boxplot pentru fiecare caracteristică numerică")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Scatter plot între două coloane relevante
plt.figure(figsize=(8, 5))
sns.scatterplot(data=selected_data, x="Hourly_Rate", y="Earnings_USD")
plt.title("🔵 Scatter plot: Hourly Rate vs Earnings")
plt.grid(True)
plt.show()
```

```python
correlation_matrix = selected_data.corr()

print("\n🔗 Corelații între coloane:")
print(correlation_matrix)

# Reprezentare grafică a corelației
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("🔗 Matricea corelațiilor")
plt.show()
```
