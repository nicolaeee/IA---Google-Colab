```python
# ✅ 1. Transformarea datelor și alegerea coloanei de ieșire (0.3 pct)

# Importăm bibliotecile necesare
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Citim datele (presupunând că sunt deja încărcate)
file_path = "/content/freelancer_earnings_bd.csv"
df = pd.read_csv(file_path, encoding="ISO-8859-1")

# Afișăm informații despre dataset înainte de transformare
print("📊 Informații despre dataset înainte de transformare:")
print(df.info())
print("\n" + "="*50 + "\n")

# Cream o copie pentru transformări
df_transformed = df.copy()

# Lista coloanelor categorice (text)
categorical_columns = ['Job_Category', 'Platform', 'Experience_Level',
                      'Client_Region', 'Payment_Method', 'Project_Type']

print("🔄 Transformarea coloanelor categorice în numerice:")

# Aplicăm Label Encoding pentru fiecare coloană categorică
label_encoders = {}
for column in categorical_columns:
    if column in df_transformed.columns:
        # Creăm un Label Encoder pentru această coloană
        le = LabelEncoder()

        # Afișăm valorile unice înainte de transformare
        print(f"\n📝 {column} - Valori unice înainte: {df_transformed[column].unique()}")

        # Aplicăm transformarea
        df_transformed[column] = le.fit_transform(df_transformed[column])

        # Salvăm encoder-ul pentru referințe viitoare
        label_encoders[column] = le

        # Afișăm valorile după transformare
        print(f"🔢 {column} - Valori după transformare: {df_transformed[column].unique()}")
        print(f"💡 Mapare: {dict(zip(le.classes_, le.transform(le.classes_)))}")

print("\n" + "="*50 + "\n")

# Selectăm coloana țintă (output) - acum Project_Type este numeric
y = df_transformed["Project_Type"]

# Selectăm toate coloanele numerice pentru input (toate caracteristicile relevante)
# Excluzem Freelancer_ID (este doar un identificator) și Project_Type (este target-ul)
feature_columns = [col for col in df_transformed.columns
                  if col not in ['Freelancer_ID', 'Project_Type']]

X = df_transformed[feature_columns]

# 🔍 Afișăm informații despre variabila țintă
print("🎯 Informații despre variabila țintă (Project_Type):")
print("🔎 Primele 10 valori din coloana de ieșire (y):")
print(y.head(10))
print(f"\n📊 Distribuția claselor în Project_Type:")
print(y.value_counts())

# 🔍 Afișăm informații despre caracteristicile de input
print(f"\n📦 Caracteristici de input (X): {list(X.columns)}")
print(f"📊 Dimensiunea X: {X.shape}")

# Verificăm că toate coloanele sunt acum numerice
print(f"\n✅ Verificarea tipurilor de date după transformare:")
print(df_transformed.dtypes)
```

```python
# ✅ 2. Împărțirea datelor în antrenare și testare (1 pct)

from sklearn.model_selection import train_test_split
import numpy as np

# Setăm seed-ul pentru reproducibilitate
np.random.seed(42)

# Verificăm dacă avem destule exemple pentru fiecare clasă
print("📊 Distribuția claselor în dataset:")
print(y.value_counts())
print(f"📊 Proporția claselor:")
print(y.value_counts(normalize=True))

# Verificăm dacă avem suficiente date pentru stratificare
min_class_count = y.value_counts().min()
print(f"\n💡 Clasa cu cele mai puține exemple: {min_class_count}")

# Împărțim în 80% antrenare și 20% testare
# Folosim stratify doar dacă avem suficiente exemple în fiecare clasă
if min_class_count >= 2:
    X_train, X_test, y_train, y_test = train_test_split(
        X, y,
        test_size=0.2,        # 20% pentru testare
        random_state=42,      # pentru rezultate reproductibile
        stratify=y            # păstrăm proporția claselor
    )
    print("✅ Stratificare aplicată cu succes!")
else:
    X_train, X_test, y_train, y_test = train_test_split(
        X, y,
        test_size=0.2,        # 20% pentru testare
        random_state=42       # pentru rezultate reproductibile
    )
    print("⚠️ Stratificare nu a fost aplicată (prea puține exemple într-o clasă)")

# 🔍 Afișăm dimensiunile seturilor rezultate
print(f"\n📊 Rezultatele împărțirii:")
print(f"📦 Dimensiune X_train: {X_train.shape}")
print(f"📦 Dimensiune X_test: {X_test.shape}")
print(f"🎯 Dimensiune y_train: {y_train.shape}")
print(f"🎯 Dimensiune y_test: {y_test.shape}")

# Verificăm distribuția claselor în seturile de antrenare și testare
print(f"\n📊 Distribuția claselor în setul de antrenare:")
print(y_train.value_counts())
print(f"📊 Distribuția claselor în setul de testare:")
print(y_test.value_counts())

# Calculăm proporțiile
print(f"\n📈 Proporții în setul de antrenare:")
print(y_train.value_counts(normalize=True))
print(f"📈 Proporții în setul de testare:")
print(y_test.value_counts(normalize=True))
```

```python
# ✅ 3. Normalizarea datelor (0.5 pct)

from sklearn.preprocessing import StandardScaler

# Cream un obiect StandardScaler pentru scalare Z-score (media 0, deviație standard 1)
scaler = StandardScaler()

# Afișăm statistici înainte de normalizare
print("📊 Statistici înainte de normalizare:")
print("Media coloanelor în X_train:")
print(X_train.mean())
print("\nDeviația standard a coloanelor în X_train:")
print(X_train.std())

# Antrenăm scalerul pe datele de antrenare și transformăm
# IMPORTANT: Antrenăm scalerul DOAR pe datele de antrenare
X_train_scaled = scaler.fit_transform(X_train)

# Folosim același scaler (deja antrenat) pentru datele de test
# IMPORTANT: Nu folosim fit_transform pe test, doar transform
X_test_scaled = scaler.transform(X_test)

# Convertim înapoi în DataFrame pentru a păstra numele coloanelor
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns, index=X_train.index)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns, index=X_test.index)

# 📊 Verificăm că normalizarea a funcționat corect
print("\n✅ Verificarea normalizării:")
print("Media după normalizare (ar trebui să fie ~0):")
print(X_train_scaled.mean().round(10))
print("\nDeviația standard după normalizare (ar trebui să fie ~1):")
print(X_train_scaled.std().round(10))

# 📊 Afișăm primele 5 rânduri din setul de antrenare normalizat
print("\n🔍 Primele 5 rânduri din X_train normalizat:")
print(X_train_scaled.head(5))

# 📊 Comparăm valorile înainte și după normalizare pentru o coloană
column_to_check = X_train.columns[0]  # Prima coloană
print(f"\n📈 Comparație pentru coloana '{column_to_check}':")
print(f"Înainte: min={X_train[column_to_check].min():.2f}, max={X_train[column_to_check].max():.2f}")
print(f"După: min={X_train_scaled[column_to_check].min():.2f}, max={X_train_scaled[column_to_check].max():.2f}")

# Verificăm dimensiunile
print(f"\n📦 Dimensiuni după normalizare:")
print(f"X_train_scaled: {X_train_scaled.shape}")
print(f"X_test_scaled: {X_test_scaled.shape}")

```

```python
# ✅ 4. Tratarea dezechilibrului claselor (SMOTE) (0.5 pct)

from imblearn.over_sampling import SMOTE
from collections import Counter

# Afișăm distribuția claselor înainte de SMOTE
print("📊 Distribuție înainte de SMOTE:")
print(f"Counter: {Counter(y_train)}")
print(f"Proporții: {y_train.value_counts(normalize=True)}")

# Calculăm câte exemple avem în total și în clasa minoritară
total_samples = len(y_train)
minority_class_count = min(Counter(y_train).values())

print(f"\n💡 Total exemple în training: {total_samples}")
print(f"💡 Clasa minoritară are: {minority_class_count} exemple")

# Determinăm numărul optim de vecini pentru SMOTE
# k_neighbors trebuie să fie mai mic decât numărul de exemple din clasa minoritară
k_neighbors = min(5, minority_class_count - 1)  # SMOTE necesită k_neighbors < numărul de exemple

if k_neighbors < 1:
    print("⚠️ Nu avem suficiente exemple pentru SMOTE. Procedăm fără balansare.")
    X_train_balanced = X_train_scaled
    y_train_balanced = y_train
else:
    print(f"🔧 Folosim k_neighbors={k_neighbors} pentru SMOTE")

    # Aplicăm SMOTE
    try:
        smote = SMOTE(random_state=42, k_neighbors=k_neighbors)
        X_train_balanced, y_train_balanced = smote.fit_resample(X_train_scaled, y_train)

        # Convertim înapoi în DataFrame
        X_train_balanced = pd.DataFrame(
            X_train_balanced,
            columns=X_train_scaled.columns
        )

        print("✅ SMOTE aplicat cu succes!")

    except Exception as e:
        print(f"⚠️ Eroare la aplicarea SMOTE: {e}")
        print("Procedăm fără balansare.")
        X_train_balanced = X_train_scaled
        y_train_balanced = y_train

# Verificăm distribuția după SMOTE
print("\n📊 Distribuție după SMOTE:")
print(f"Counter: {Counter(y_train_balanced)}")

if hasattr(y_train_balanced, 'value_counts'):
    print(f"Proporții: {y_train_balanced.value_counts(normalize=True)}")
else:
    # Dacă y_train_balanced este numpy array
    unique, counts = np.unique(y_train_balanced, return_counts=True)
    proportions = counts / len(y_train_balanced)
    for val, prop in zip(unique, proportions):
        print(f"Clasa {val}: {prop:.3f}")

# Afișăm câte exemple noi au fost generate
original_size = len(y_train)
new_size = len(y_train_balanced)
print(f"\n📈 Dimensiune înainte de SMOTE: {original_size}")
print(f"📈 Dimensiune după SMOTE: {new_size}")
print(f"📈 Exemple noi generate: {new_size - original_size}")

# Verificăm dimensiunile finale
print(f"\n📦 Dimensiuni finale:")
print(f"X_train_balanced: {X_train_balanced.shape}")
print(f"y_train_balanced: {len(y_train_balanced)}")
```

```python
# ✅ 5. Selectarea celor mai bune caracteristici (0.7 pct)

from sklearn.feature_selection import SelectKBest, f_classif
import matplotlib.pyplot as plt

# Afișăm toate caracteristicile disponibile
print("📊 Caracteristicile disponibile:")
for i, feature in enumerate(X_train_balanced.columns):
    print(f"{i+1}. {feature}")

# Determinăm numărul optim de caracteristici
total_features = X_train_balanced.shape[1]
# Alegem cel mult 75% din caracteristici, dar cel puțin 3
k_features = min(max(3, int(total_features * 0.75)), total_features)

print(f"\n🎯 Vom selecta {k_features} caracteristici din {total_features} disponibile")

# Creăm selectorul de caracteristici
# f_classif calculează F-statistica pentru fiecare caracteristică
selector = SelectKBest(score_func=f_classif, k=k_features)

# Aplicăm selecția pe datele de antrenare balansate
X_train_selected = selector.fit_transform(X_train_balanced, y_train_balanced)

# Aplicăm aceeași transformare pe datele de test
X_test_selected = selector.transform(X_test_scaled)

# Obținem numele caracteristicilor selectate
selected_features_mask = selector.get_support()  # Array boolean
selected_features = X_train_balanced.columns[selected_features_mask]

print("\n✅ Caracteristici selectate:")
for i, feature in enumerate(selected_features):
    print(f"{i+1}. {feature}")

# Obținem scorurile pentru toate caracteristicile
feature_scores = selector.scores_
feature_pvalues = selector.pvalues_

# Creăm un DataFrame cu scorurile pentru toate caracteristicile
feature_importance_df = pd.DataFrame({
    'Feature': X_train_balanced.columns,
    'Score': feature_scores,
    'P_value': feature_pvalues,
    'Selected': selected_features_mask
})

# Sortăm după scor
feature_importance_df = feature_importance_df.sort_values('Score', ascending=False)

print("\n📊 Scorurile tuturor caracteristicilor (sortate descrescător):")
print(feature_importance_df)

# Vizualizăm importanța caracteristicilor
plt.figure(figsize=(12, 6))
colors = ['green' if selected else 'lightgray' for selected in feature_importance_df['Selected']]
plt.bar(range(len(feature_importance_df)), feature_importance_df['Score'], color=colors)
plt.xlabel('Caracteristici')
plt.ylabel('Scor F-statistic')
plt.title(f'Importanța Caracteristicilor (Verde = Selectate, primele {k_features})')
plt.xticks(range(len(feature_importance_df)), feature_importance_df['Feature'], rotation=45)
plt.tight_layout()
plt.grid(axis='y', alpha=0.3)
plt.show()

# Convertim rezultatele în DataFrames pentru a păstra numele coloanelor
X_train_selected = pd.DataFrame(
    X_train_selected,
    columns=selected_features,
    index=X_train_balanced.index if hasattr(X_train_balanced, 'index') else None
)

X_test_selected = pd.DataFrame(
    X_test_selected,
    columns=selected_features,
    index=X_test_scaled.index if hasattr(X_test_scaled, 'index') else None
)

# Afișăm dimensiunile finale
print(f"\n📦 Dimensiuni după selecția caracteristicilor:")
print(f"X_train_selected: {X_train_selected.shape}")
print(f"X_test_selected: {X_test_selected.shape}")

# Afișăm primele rânduri din setul cu caracteristici selectate
print(f"\n🔍 Primele 3 rânduri cu caracteristici selectate:")
print(X_train_selected.head(3))
```

```python
# ✅ 6. Crearea modelului (rețea neuronală) (1 pct)

from sklearn.neural_network import MLPClassifier
import numpy as np

# Afișăm informații despre datele de antrenare
print("📊 Informații despre setul de antrenare final:")
print(f"Forma X_train_selected: {X_train_selected.shape}")
print(f"Numărul de caracteristici: {X_train_selected.shape[1]}")
print(f"Numărul de exemple: {X_train_selected.shape[0]}")
print(f"Numărul de clase: {len(np.unique(y_train_balanced))}")

# Determinăm arhitectura rețelei neuronale în funcție de dimensiunea datelor
n_features = X_train_selected.shape[1]
n_samples = X_train_selected.shape[0]

# Calculăm dimensiunea straturilor ascunse
# Prima strat: de obicei între n_features și 2*n_features
# Al doilea strat: mai mic, între n_features/2 și n_features
hidden_layer_1 = min(max(10, n_features * 2), 100)  # Limităm la max 100 neuroni
hidden_layer_2 = min(max(5, n_features), 50)        # Limităm la max 50 neuroni

print(f"\n🧠 Arhitectura rețelei neuronale selectate:")
print(f"Stratul 1 (ascuns): {hidden_layer_1} neuroni")
print(f"Stratul 2 (ascuns): {hidden_layer_2} neuroni")

# Cream modelul de rețea neuronală (MLP - Multi-Layer Perceptron)
model = MLPClassifier(
    hidden_layer_sizes=(hidden_layer_1, hidden_layer_2),  # Două straturi ascunse
    activation='relu',           # Funcția de activare ReLU (Rectified Linear Unit)
                                # ReLU: f(x) = max(0, x) - simplu și eficient
    solver='adam',               # Optimizatorul Adam - adaptiv și eficient
                                # Adam combină momentul și rata de învățare adaptivă
    alpha=0.0001,               # Regularizare L2 pentru a preveni overfitting-ul
    batch_size='auto',          # Dimensiunea batch-ului calculată automat
    learning_rate='constant',    # Rata de învățare constantă
    learning_rate_init=0.001,   # Rata de învățare inițială
    max_iter=1000,              # Numărul maxim de epoci (iterații)
    shuffle=True,               # Amestecă datele la fiecare epocă
    random_state=42,            # Pentru rezultate reproductibile
    verbose=True,               # Afișează progresul antrenării
    early_stopping=True,        # Oprește antrenarea dacă nu mai îmbunătățește
    validation_fraction=0.1,    # 10% din datele de antrenare pentru validare
    n_iter_no_change=10         # Oprește dacă nu se îmbunătățește 10 iterații consecutive
)

print("\n🔧 Parametrii modelului:")
print(f"Funcție activare: {model.activation}")
print(f"Solver: {model.solver}")
print(f"Regularizare (alpha): {model.alpha}")
print(f"Rata învățare: {model.learning_rate_init}")
print(f"Max iterații: {model.max_iter}")
print(f"Early stopping: {model.early_stopping}")

# Antrenăm modelul pe datele balansate și cu caracteristicile selectate
print("\n🚀 Începem antrenarea modelului...")
print("(Acest proces poate dura câteva secunde/minute)")

try:
    model.fit(X_train_selected, y_train_balanced)
    print("\n✅ Modelul a fost antrenat cu succes!")

    # Afișăm informații despre antrenare
    print(f"📊 Numărul de iterații efectuate: {model.n_iter_}")
    print(f"📊 Funcția de cost finală: {model.loss_:.6f}")

    # Afișăm dimensiunile straturilor din rețea
    print(f"\n🔢 Dimensiunile straturilor în rețeaua antrenată:")
    for i, layer_size in enumerate(model.coefs_):
        print(f"Stratul {i} -> Stratul {i+1}: {layer_size.shape}")

except Exception as e:
    print(f"❌ Eroare la antrenarea modelului: {e}")
    print("Încearcă să reduci dimensiunea rețelei sau să modifici parametrii.")
```

```python
# ✅ 7. Acuratețea pe setul de antrenare (1 pct)

from sklearn.metrics import accuracy_score, classification_report
import numpy as np

# Verificăm dacă modelul a fost antrenat cu succes
if hasattr(model, 'coefs_'):
    print("✅ Modelul este antrenat și gata pentru evaluare")
else:
    print("❌ Modelul nu pare să fie antrenat corect")

# Facem predicții pe setul de antrenare
print("🔮 Facem predicții pe setul de antrenare...")
y_train_pred = model.predict(X_train_selected)

# Calculăm acuratețea pe setul de antrenare
train_accuracy = accuracy_score(y_train_balanced, y_train_pred)

print(f"\n🎯 ACURATEȚE PE SETUL DE ANTRENARE: {train_accuracy:.3f} ({train_accuracy*100:.1f}%)")

# Afișăm informații detaliate despre performanță
print("\n📊 Analiza detaliată a performanței pe antrenare:")

# Calculăm și afișăm distribuția predicțiilor
pred_distribution = pd.Series(y_train_pred).value_counts().sort_index()
true_distribution = pd.Series(y_train_balanced).value_counts().sort_index()

print("📈 Distribuția claselor:")
print("Clasa | Adevărat | Prezis")
print("-" * 25)
for class_label in sorted(np.unique(y_train_balanced)):
    true_count = true_distribution.get(class_label, 0)
    pred_count = pred_distribution.get(class_label, 0)
    print(f"  {class_label}   |    {true_count:3d}    |  {pred_count:3d}")

# Calculăm acuratețea pentru fiecare clasă
print("\n🎯 Acuratețea pentru fiecare clasă:")
for class_label in sorted(np.unique(y_train_balanced)):
    # Găsim indicii pentru această clasă
    class_indices = np.where(y_train_balanced == class_label)[0]

    if len(class_indices) > 0:
        # Calculăm acuratețea pentru această clasă
        class_true = y_train_balanced[class_indices]
        class_pred = y_train_pred[class_indices]
        class_accuracy = accuracy_score(class_true, class_pred)

        print(f"Clasa {class_label}: {class_accuracy:.3f} ({class_accuracy*100:.1f}%)")

# Calculăm câte predicții corecte avem în total
correct_predictions = np.sum(y_train_pred == y_train_balanced)
total_predictions = len(y_train_balanced)

print(f"\n📊 Rezumat:")
print(f"Total exemple: {total_predictions}")
print(f"Predicții corecte: {correct_predictions}")
print(f"Predicții greșite: {total_predictions - correct_predictions}")
print(f"Rata de eroare: {(1 - train_accuracy):.3f} ({(1 - train_accuracy)*100:.1f}%)")

# Verificăm dacă există overfitting (acuratețe foarte mare pe antrenare)
if train_accuracy > 0.95:
    print("\n⚠️  ATENȚIE: Acuratețea pe antrenare este foarte mare (>95%)")
    print("    Acest lucru poate indica overfitting.")
    print("    Verifică acuratețea pe setul de testare pentru confirmare.")
elif train_accuracy < 0.6:
    print("\n⚠️  ATENȚIE: Acuratețea pe antrenare este scăzută (<60%)")
    print("    Modelul ar putea avea nevoie de:")
    print("    - Mai multe iterații")
    print("    - O arhitectură diferită")
    print("    - Mai multe date de antrenare")
else:
    print(f"\n✅ Acuratețea pe antrenare pare rezonabilă ({train_accuracy*100:.1f}%)")
```
