```python
import pandas as pd
import matplotlib.pyplot as plt

# Citirea fișierului CSV
file_path = "/content/freelancer_earnings_bd.csv"  # Asigură-te că încarci fișierul în Colab
df = pd.read_csv(file_path, encoding="ISO-8859-1")

# Agregare a veniturilor pe categorie de job
df_grouped = df.groupby("Job_Category")["Earnings_USD"].sum().reset_index()

# Sortare și selecție top 10 categorii
df_sorted = df_grouped.sort_values(by="Earnings_USD", ascending=False)[:10]

# Creare grafic
plt.figure(figsize=(12, 6))
plt.bar(df_sorted["Job_Category"], df_sorted["Earnings_USD"], color='royalblue')
plt.xticks(rotation=45, ha="right")
plt.ylabel("Venit total (USD)")
plt.title("Top 10 categorii de joburi cu cele mai mari venituri")
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Afișare grafic
plt.show()
```

```python
import pandas as pd

def load_data(file_path):
    """
    Funcție pentru citirea fișierului CSV.
    """
    return pd.read_csv(file_path, encoding="ISO-8859-1")

# Exemplu de utilizare
if __name__ == "__main__":
    file_path = "/content/freelancer_earnings_bd.csv"  # Încarcă fișierul în Google Colab
    df = load_data(file_path)
    print(df.head())
```

```python
import matplotlib.pyplot as plt


def plot_stock_chart(df):
    df_sorted = df.sort_values(by="Earnings_USD", ascending=False)[:20]
    plt.figure(figsize=(12, 6))
    plt.vlines(df_sorted["Freelancer_ID"], df_sorted["Earnings_USD"] * 0.8, df_sorted["Earnings_USD"] * 1.2, color='black')
    plt.scatter(df_sorted["Freelancer_ID"], df_sorted["Earnings_USD"], color='green', label="Venituri")
    plt.xlabel("Freelancer ID")
    plt.ylabel("Venituri (USD)")
    plt.title("Variația veniturilor freelancerilor (stil bursă)")
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

if __name__ == "__main__":
    df = load_data("/content/freelancer_earnings_bd.csv")
    plot_stock_chart(df)
```
