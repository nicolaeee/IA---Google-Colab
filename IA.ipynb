```python
# 九 1. Transformarea datelor 탳i alegerea coloanei de ie탳ire (0.3 pct)

# Import캒m bibliotecile necesare
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Citim datele (presupun칙nd c캒 sunt deja 칥nc캒rcate)
file_path = "/content/freelancer_earnings_bd.csv"
df = pd.read_csv(file_path, encoding="ISO-8859-1")

# Afi탳캒m informa탵ii despre dataset 칥nainte de transformare
print("游늵 Informa탵ii despre dataset 칥nainte de transformare:")
print(df.info())
print("\n" + "="*50 + "\n")

# Cream o copie pentru transform캒ri
df_transformed = df.copy()

# Lista coloanelor categorice (text)
categorical_columns = ['Job_Category', 'Platform', 'Experience_Level',
                      'Client_Region', 'Payment_Method', 'Project_Type']

print("游댃 Transformarea coloanelor categorice 칥n numerice:")

# Aplic캒m Label Encoding pentru fiecare coloan캒 categoric캒
label_encoders = {}
for column in categorical_columns:
    if column in df_transformed.columns:
        # Cre캒m un Label Encoder pentru aceast캒 coloan캒
        le = LabelEncoder()

        # Afi탳캒m valorile unice 칥nainte de transformare
        print(f"\n游닇 {column} - Valori unice 칥nainte: {df_transformed[column].unique()}")

        # Aplic캒m transformarea
        df_transformed[column] = le.fit_transform(df_transformed[column])

        # Salv캒m encoder-ul pentru referin탵e viitoare
        label_encoders[column] = le

        # Afi탳캒m valorile dup캒 transformare
        print(f"游댝 {column} - Valori dup캒 transformare: {df_transformed[column].unique()}")
        print(f"游눠 Mapare: {dict(zip(le.classes_, le.transform(le.classes_)))}")

print("\n" + "="*50 + "\n")

# Select캒m coloana 탵int캒 (output) - acum Project_Type este numeric
y = df_transformed["Project_Type"]

# Select캒m toate coloanele numerice pentru input (toate caracteristicile relevante)
# Excluzem Freelancer_ID (este doar un identificator) 탳i Project_Type (este target-ul)
feature_columns = [col for col in df_transformed.columns
                  if col not in ['Freelancer_ID', 'Project_Type']]

X = df_transformed[feature_columns]

# 游댌 Afi탳캒m informa탵ii despre variabila 탵int캒
print("游꿢 Informa탵ii despre variabila 탵int캒 (Project_Type):")
print("游댍 Primele 10 valori din coloana de ie탳ire (y):")
print(y.head(10))
print(f"\n游늵 Distribu탵ia claselor 칥n Project_Type:")
print(y.value_counts())

# 游댌 Afi탳캒m informa탵ii despre caracteristicile de input
print(f"\n游닍 Caracteristici de input (X): {list(X.columns)}")
print(f"游늵 Dimensiunea X: {X.shape}")

# Verific캒m c캒 toate coloanele sunt acum numerice
print(f"\n九 Verificarea tipurilor de date dup캒 transformare:")
print(df_transformed.dtypes)
```

```python
# 九 2. 칉mp캒r탵irea datelor 칥n antrenare 탳i testare (1 pct)

from sklearn.model_selection import train_test_split
import numpy as np

# Set캒m seed-ul pentru reproducibilitate
np.random.seed(42)

# Verific캒m dac캒 avem destule exemple pentru fiecare clas캒
print("游늵 Distribu탵ia claselor 칥n dataset:")
print(y.value_counts())
print(f"游늵 Propor탵ia claselor:")
print(y.value_counts(normalize=True))

# Verific캒m dac캒 avem suficiente date pentru stratificare
min_class_count = y.value_counts().min()
print(f"\n游눠 Clasa cu cele mai pu탵ine exemple: {min_class_count}")

# 칉mp캒r탵im 칥n 80% antrenare 탳i 20% testare
# Folosim stratify doar dac캒 avem suficiente exemple 칥n fiecare clas캒
if min_class_count >= 2:
    X_train, X_test, y_train, y_test = train_test_split(
        X, y,
        test_size=0.2,        # 20% pentru testare
        random_state=42,      # pentru rezultate reproductibile
        stratify=y            # p캒str캒m propor탵ia claselor
    )
    print("九 Stratificare aplicat캒 cu succes!")
else:
    X_train, X_test, y_train, y_test = train_test_split(
        X, y,
        test_size=0.2,        # 20% pentru testare
        random_state=42       # pentru rezultate reproductibile
    )
    print("丘멆잺 Stratificare nu a fost aplicat캒 (prea pu탵ine exemple 칥ntr-o clas캒)")

# 游댌 Afi탳캒m dimensiunile seturilor rezultate
print(f"\n游늵 Rezultatele 칥mp캒r탵irii:")
print(f"游닍 Dimensiune X_train: {X_train.shape}")
print(f"游닍 Dimensiune X_test: {X_test.shape}")
print(f"游꿢 Dimensiune y_train: {y_train.shape}")
print(f"游꿢 Dimensiune y_test: {y_test.shape}")

# Verific캒m distribu탵ia claselor 칥n seturile de antrenare 탳i testare
print(f"\n游늵 Distribu탵ia claselor 칥n setul de antrenare:")
print(y_train.value_counts())
print(f"游늵 Distribu탵ia claselor 칥n setul de testare:")
print(y_test.value_counts())

# Calcul캒m propor탵iile
print(f"\n游늳 Propor탵ii 칥n setul de antrenare:")
print(y_train.value_counts(normalize=True))
print(f"游늳 Propor탵ii 칥n setul de testare:")
print(y_test.value_counts(normalize=True))
```

```python
# 九 3. Normalizarea datelor (0.5 pct)

from sklearn.preprocessing import StandardScaler

# Cream un obiect StandardScaler pentru scalare Z-score (media 0, devia탵ie standard 1)
scaler = StandardScaler()

# Afi탳캒m statistici 칥nainte de normalizare
print("游늵 Statistici 칥nainte de normalizare:")
print("Media coloanelor 칥n X_train:")
print(X_train.mean())
print("\nDevia탵ia standard a coloanelor 칥n X_train:")
print(X_train.std())

# Antren캒m scalerul pe datele de antrenare 탳i transform캒m
# IMPORTANT: Antren캒m scalerul DOAR pe datele de antrenare
X_train_scaled = scaler.fit_transform(X_train)

# Folosim acela탳i scaler (deja antrenat) pentru datele de test
# IMPORTANT: Nu folosim fit_transform pe test, doar transform
X_test_scaled = scaler.transform(X_test)

# Convertim 칥napoi 칥n DataFrame pentru a p캒stra numele coloanelor
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns, index=X_train.index)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns, index=X_test.index)

# 游늵 Verific캒m c캒 normalizarea a func탵ionat corect
print("\n九 Verificarea normaliz캒rii:")
print("Media dup캒 normalizare (ar trebui s캒 fie ~0):")
print(X_train_scaled.mean().round(10))
print("\nDevia탵ia standard dup캒 normalizare (ar trebui s캒 fie ~1):")
print(X_train_scaled.std().round(10))

# 游늵 Afi탳캒m primele 5 r칙nduri din setul de antrenare normalizat
print("\n游댌 Primele 5 r칙nduri din X_train normalizat:")
print(X_train_scaled.head(5))

# 游늵 Compar캒m valorile 칥nainte 탳i dup캒 normalizare pentru o coloan캒
column_to_check = X_train.columns[0]  # Prima coloan캒
print(f"\n游늳 Compara탵ie pentru coloana '{column_to_check}':")
print(f"칉nainte: min={X_train[column_to_check].min():.2f}, max={X_train[column_to_check].max():.2f}")
print(f"Dup캒: min={X_train_scaled[column_to_check].min():.2f}, max={X_train_scaled[column_to_check].max():.2f}")

# Verific캒m dimensiunile
print(f"\n游닍 Dimensiuni dup캒 normalizare:")
print(f"X_train_scaled: {X_train_scaled.shape}")
print(f"X_test_scaled: {X_test_scaled.shape}")

```

```python
# 九 4. Tratarea dezechilibrului claselor (SMOTE) (0.5 pct)

from imblearn.over_sampling import SMOTE
from collections import Counter

# Afi탳캒m distribu탵ia claselor 칥nainte de SMOTE
print("游늵 Distribu탵ie 칥nainte de SMOTE:")
print(f"Counter: {Counter(y_train)}")
print(f"Propor탵ii: {y_train.value_counts(normalize=True)}")

# Calcul캒m c칙te exemple avem 칥n total 탳i 칥n clasa minoritar캒
total_samples = len(y_train)
minority_class_count = min(Counter(y_train).values())

print(f"\n游눠 Total exemple 칥n training: {total_samples}")
print(f"游눠 Clasa minoritar캒 are: {minority_class_count} exemple")

# Determin캒m num캒rul optim de vecini pentru SMOTE
# k_neighbors trebuie s캒 fie mai mic dec칙t num캒rul de exemple din clasa minoritar캒
k_neighbors = min(5, minority_class_count - 1)  # SMOTE necesit캒 k_neighbors < num캒rul de exemple

if k_neighbors < 1:
    print("丘멆잺 Nu avem suficiente exemple pentru SMOTE. Proced캒m f캒r캒 balansare.")
    X_train_balanced = X_train_scaled
    y_train_balanced = y_train
else:
    print(f"游댢 Folosim k_neighbors={k_neighbors} pentru SMOTE")

    # Aplic캒m SMOTE
    try:
        smote = SMOTE(random_state=42, k_neighbors=k_neighbors)
        X_train_balanced, y_train_balanced = smote.fit_resample(X_train_scaled, y_train)

        # Convertim 칥napoi 칥n DataFrame
        X_train_balanced = pd.DataFrame(
            X_train_balanced,
            columns=X_train_scaled.columns
        )

        print("九 SMOTE aplicat cu succes!")

    except Exception as e:
        print(f"丘멆잺 Eroare la aplicarea SMOTE: {e}")
        print("Proced캒m f캒r캒 balansare.")
        X_train_balanced = X_train_scaled
        y_train_balanced = y_train

# Verific캒m distribu탵ia dup캒 SMOTE
print("\n游늵 Distribu탵ie dup캒 SMOTE:")
print(f"Counter: {Counter(y_train_balanced)}")

if hasattr(y_train_balanced, 'value_counts'):
    print(f"Propor탵ii: {y_train_balanced.value_counts(normalize=True)}")
else:
    # Dac캒 y_train_balanced este numpy array
    unique, counts = np.unique(y_train_balanced, return_counts=True)
    proportions = counts / len(y_train_balanced)
    for val, prop in zip(unique, proportions):
        print(f"Clasa {val}: {prop:.3f}")

# Afi탳캒m c칙te exemple noi au fost generate
original_size = len(y_train)
new_size = len(y_train_balanced)
print(f"\n游늳 Dimensiune 칥nainte de SMOTE: {original_size}")
print(f"游늳 Dimensiune dup캒 SMOTE: {new_size}")
print(f"游늳 Exemple noi generate: {new_size - original_size}")

# Verific캒m dimensiunile finale
print(f"\n游닍 Dimensiuni finale:")
print(f"X_train_balanced: {X_train_balanced.shape}")
print(f"y_train_balanced: {len(y_train_balanced)}")
```

```python
# 九 5. Selectarea celor mai bune caracteristici (0.7 pct)

from sklearn.feature_selection import SelectKBest, f_classif
import matplotlib.pyplot as plt

# Afi탳캒m toate caracteristicile disponibile
print("游늵 Caracteristicile disponibile:")
for i, feature in enumerate(X_train_balanced.columns):
    print(f"{i+1}. {feature}")

# Determin캒m num캒rul optim de caracteristici
total_features = X_train_balanced.shape[1]
# Alegem cel mult 75% din caracteristici, dar cel pu탵in 3
k_features = min(max(3, int(total_features * 0.75)), total_features)

print(f"\n游꿢 Vom selecta {k_features} caracteristici din {total_features} disponibile")

# Cre캒m selectorul de caracteristici
# f_classif calculeaz캒 F-statistica pentru fiecare caracteristic캒
selector = SelectKBest(score_func=f_classif, k=k_features)

# Aplic캒m selec탵ia pe datele de antrenare balansate
X_train_selected = selector.fit_transform(X_train_balanced, y_train_balanced)

# Aplic캒m aceea탳i transformare pe datele de test
X_test_selected = selector.transform(X_test_scaled)

# Ob탵inem numele caracteristicilor selectate
selected_features_mask = selector.get_support()  # Array boolean
selected_features = X_train_balanced.columns[selected_features_mask]

print("\n九 Caracteristici selectate:")
for i, feature in enumerate(selected_features):
    print(f"{i+1}. {feature}")

# Ob탵inem scorurile pentru toate caracteristicile
feature_scores = selector.scores_
feature_pvalues = selector.pvalues_

# Cre캒m un DataFrame cu scorurile pentru toate caracteristicile
feature_importance_df = pd.DataFrame({
    'Feature': X_train_balanced.columns,
    'Score': feature_scores,
    'P_value': feature_pvalues,
    'Selected': selected_features_mask
})

# Sort캒m dup캒 scor
feature_importance_df = feature_importance_df.sort_values('Score', ascending=False)

print("\n游늵 Scorurile tuturor caracteristicilor (sortate descresc캒tor):")
print(feature_importance_df)

# Vizualiz캒m importan탵a caracteristicilor
plt.figure(figsize=(12, 6))
colors = ['green' if selected else 'lightgray' for selected in feature_importance_df['Selected']]
plt.bar(range(len(feature_importance_df)), feature_importance_df['Score'], color=colors)
plt.xlabel('Caracteristici')
plt.ylabel('Scor F-statistic')
plt.title(f'Importan탵a Caracteristicilor (Verde = Selectate, primele {k_features})')
plt.xticks(range(len(feature_importance_df)), feature_importance_df['Feature'], rotation=45)
plt.tight_layout()
plt.grid(axis='y', alpha=0.3)
plt.show()

# Convertim rezultatele 칥n DataFrames pentru a p캒stra numele coloanelor
X_train_selected = pd.DataFrame(
    X_train_selected,
    columns=selected_features,
    index=X_train_balanced.index if hasattr(X_train_balanced, 'index') else None
)

X_test_selected = pd.DataFrame(
    X_test_selected,
    columns=selected_features,
    index=X_test_scaled.index if hasattr(X_test_scaled, 'index') else None
)

# Afi탳캒m dimensiunile finale
print(f"\n游닍 Dimensiuni dup캒 selec탵ia caracteristicilor:")
print(f"X_train_selected: {X_train_selected.shape}")
print(f"X_test_selected: {X_test_selected.shape}")

# Afi탳캒m primele r칙nduri din setul cu caracteristici selectate
print(f"\n游댌 Primele 3 r칙nduri cu caracteristici selectate:")
print(X_train_selected.head(3))
```

```python
# 九 6. Crearea modelului (re탵ea neuronal캒) (1 pct)

from sklearn.neural_network import MLPClassifier
import numpy as np

# Afi탳캒m informa탵ii despre datele de antrenare
print("游늵 Informa탵ii despre setul de antrenare final:")
print(f"Forma X_train_selected: {X_train_selected.shape}")
print(f"Num캒rul de caracteristici: {X_train_selected.shape[1]}")
print(f"Num캒rul de exemple: {X_train_selected.shape[0]}")
print(f"Num캒rul de clase: {len(np.unique(y_train_balanced))}")

# Determin캒m arhitectura re탵elei neuronale 칥n func탵ie de dimensiunea datelor
n_features = X_train_selected.shape[1]
n_samples = X_train_selected.shape[0]

# Calcul캒m dimensiunea straturilor ascunse
# Prima strat: de obicei 칥ntre n_features 탳i 2*n_features
# Al doilea strat: mai mic, 칥ntre n_features/2 탳i n_features
hidden_layer_1 = min(max(10, n_features * 2), 100)  # Limit캒m la max 100 neuroni
hidden_layer_2 = min(max(5, n_features), 50)        # Limit캒m la max 50 neuroni

print(f"\n游 Arhitectura re탵elei neuronale selectate:")
print(f"Stratul 1 (ascuns): {hidden_layer_1} neuroni")
print(f"Stratul 2 (ascuns): {hidden_layer_2} neuroni")

# Cream modelul de re탵ea neuronal캒 (MLP - Multi-Layer Perceptron)
model = MLPClassifier(
    hidden_layer_sizes=(hidden_layer_1, hidden_layer_2),  # Dou캒 straturi ascunse
    activation='relu',           # Func탵ia de activare ReLU (Rectified Linear Unit)
                                # ReLU: f(x) = max(0, x) - simplu 탳i eficient
    solver='adam',               # Optimizatorul Adam - adaptiv 탳i eficient
                                # Adam combin캒 momentul 탳i rata de 칥nv캒탵are adaptiv캒
    alpha=0.0001,               # Regularizare L2 pentru a preveni overfitting-ul
    batch_size='auto',          # Dimensiunea batch-ului calculat캒 automat
    learning_rate='constant',    # Rata de 칥nv캒탵are constant캒
    learning_rate_init=0.001,   # Rata de 칥nv캒탵are ini탵ial캒
    max_iter=1000,              # Num캒rul maxim de epoci (itera탵ii)
    shuffle=True,               # Amestec캒 datele la fiecare epoc캒
    random_state=42,            # Pentru rezultate reproductibile
    verbose=True,               # Afi탳eaz캒 progresul antren캒rii
    early_stopping=True,        # Opre탳te antrenarea dac캒 nu mai 칥mbun캒t캒탵e탳te
    validation_fraction=0.1,    # 10% din datele de antrenare pentru validare
    n_iter_no_change=10         # Opre탳te dac캒 nu se 칥mbun캒t캒탵e탳te 10 itera탵ii consecutive
)

print("\n游댢 Parametrii modelului:")
print(f"Func탵ie activare: {model.activation}")
print(f"Solver: {model.solver}")
print(f"Regularizare (alpha): {model.alpha}")
print(f"Rata 칥nv캒탵are: {model.learning_rate_init}")
print(f"Max itera탵ii: {model.max_iter}")
print(f"Early stopping: {model.early_stopping}")

# Antren캒m modelul pe datele balansate 탳i cu caracteristicile selectate
print("\n游 칉ncepem antrenarea modelului...")
print("(Acest proces poate dura c칙teva secunde/minute)")

try:
    model.fit(X_train_selected, y_train_balanced)
    print("\n九 Modelul a fost antrenat cu succes!")

    # Afi탳캒m informa탵ii despre antrenare
    print(f"游늵 Num캒rul de itera탵ii efectuate: {model.n_iter_}")
    print(f"游늵 Func탵ia de cost final캒: {model.loss_:.6f}")

    # Afi탳캒m dimensiunile straturilor din re탵ea
    print(f"\n游댝 Dimensiunile straturilor 칥n re탵eaua antrenat캒:")
    for i, layer_size in enumerate(model.coefs_):
        print(f"Stratul {i} -> Stratul {i+1}: {layer_size.shape}")

except Exception as e:
    print(f"仇 Eroare la antrenarea modelului: {e}")
    print("칉ncearc캒 s캒 reduci dimensiunea re탵elei sau s캒 modifici parametrii.")
```

```python
# 九 7. Acurate탵ea pe setul de antrenare (1 pct)

from sklearn.metrics import accuracy_score, classification_report
import numpy as np

# Verific캒m dac캒 modelul a fost antrenat cu succes
if hasattr(model, 'coefs_'):
    print("九 Modelul este antrenat 탳i gata pentru evaluare")
else:
    print("仇 Modelul nu pare s캒 fie antrenat corect")

# Facem predic탵ii pe setul de antrenare
print("游댩 Facem predic탵ii pe setul de antrenare...")
y_train_pred = model.predict(X_train_selected)

# Calcul캒m acurate탵ea pe setul de antrenare
train_accuracy = accuracy_score(y_train_balanced, y_train_pred)

print(f"\n游꿢 ACURATE탴E PE SETUL DE ANTRENARE: {train_accuracy:.3f} ({train_accuracy*100:.1f}%)")

# Afi탳캒m informa탵ii detaliate despre performan탵캒
print("\n游늵 Analiza detaliat캒 a performan탵ei pe antrenare:")

# Calcul캒m 탳i afi탳캒m distribu탵ia predic탵iilor
pred_distribution = pd.Series(y_train_pred).value_counts().sort_index()
true_distribution = pd.Series(y_train_balanced).value_counts().sort_index()

print("游늳 Distribu탵ia claselor:")
print("Clasa | Adev캒rat | Prezis")
print("-" * 25)
for class_label in sorted(np.unique(y_train_balanced)):
    true_count = true_distribution.get(class_label, 0)
    pred_count = pred_distribution.get(class_label, 0)
    print(f"  {class_label}   |    {true_count:3d}    |  {pred_count:3d}")

# Calcul캒m acurate탵ea pentru fiecare clas캒
print("\n游꿢 Acurate탵ea pentru fiecare clas캒:")
for class_label in sorted(np.unique(y_train_balanced)):
    # G캒sim indicii pentru aceast캒 clas캒
    class_indices = np.where(y_train_balanced == class_label)[0]

    if len(class_indices) > 0:
        # Calcul캒m acurate탵ea pentru aceast캒 clas캒
        class_true = y_train_balanced[class_indices]
        class_pred = y_train_pred[class_indices]
        class_accuracy = accuracy_score(class_true, class_pred)

        print(f"Clasa {class_label}: {class_accuracy:.3f} ({class_accuracy*100:.1f}%)")

# Calcul캒m c칙te predic탵ii corecte avem 칥n total
correct_predictions = np.sum(y_train_pred == y_train_balanced)
total_predictions = len(y_train_balanced)

print(f"\n游늵 Rezumat:")
print(f"Total exemple: {total_predictions}")
print(f"Predic탵ii corecte: {correct_predictions}")
print(f"Predic탵ii gre탳ite: {total_predictions - correct_predictions}")
print(f"Rata de eroare: {(1 - train_accuracy):.3f} ({(1 - train_accuracy)*100:.1f}%)")

# Verific캒m dac캒 exist캒 overfitting (acurate탵e foarte mare pe antrenare)
if train_accuracy > 0.95:
    print("\n丘멆잺  ATEN탴IE: Acurate탵ea pe antrenare este foarte mare (>95%)")
    print("    Acest lucru poate indica overfitting.")
    print("    Verific캒 acurate탵ea pe setul de testare pentru confirmare.")
elif train_accuracy < 0.6:
    print("\n丘멆잺  ATEN탴IE: Acurate탵ea pe antrenare este sc캒zut캒 (<60%)")
    print("    Modelul ar putea avea nevoie de:")
    print("    - Mai multe itera탵ii")
    print("    - O arhitectur캒 diferit캒")
    print("    - Mai multe date de antrenare")
else:
    print(f"\n九 Acurate탵ea pe antrenare pare rezonabil캒 ({train_accuracy*100:.1f}%)")
```
